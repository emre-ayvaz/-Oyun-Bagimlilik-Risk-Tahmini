# 1. GIRIS


## 1.1 Proje Basligi


**Dijital Oyun Kullaniminin Bireylerin Psikolojik ve Sosyolojik Durumlari Uzerindeki Etkisinin Analizi: Bir Makine Ogrenmesi Yaklasimi**


## 1.2 Projenin Konusu ve Amaci


Bu calisma, dijital oyun aliskanliklarinin (gunluk oynama suresi, oyun turu, kullanilan platform vb.) bireylerin ruh hali, uyku kalitesi ve sosyal izolasyon duzeyleri uzerindeki etkilerini veri odakli yontemlerle incelemeyi amaclamaktadir.

Projenin temel hedefi; demografik ve davranissal degiskenlerden yola cikarak bireylerin **oyun bagimlilik risk seviyesini** (gaming addiction risk level) tahmin eden, yorumlanabilir ve genellenebilir bir makine ogrenmesi modeli gelistirmektir. Bu amacla K-En Yakin Komsu (KNN), Rastgele Orman (Random Forest) ve Lojistik Regresyon algoritmalari karsilastirmali olarak uygulanmistir.

Calismanin onemi; bagimlilik riskinin erken ve nesnel bicimde tespit edilmesiyle onleyici mudahale olanaklarina zemin hazirlamasindan kaynaklanmaktadir. Veri temelli bir risk siniflandirmasi, klinik degerlendirmelere destek niteligini tasiyabilir.


## 1.3 Problem Tanimi


Dijital oyunlarin kontrolsuz kullanimi; uyku bozukluklari, akademik performans dususu, sosyal izolasyon ve fiziksel agrilar gibi cok boyutlu olumsuz sonuclara yol acabilidir. Dunya Saglik Orgutu (WHO), 2018 yilinda oyun bozuklugunu (gaming disorder) resmi olarak ICD-11 kapsamina almis; bu durum konuyu kuresel bir halk sagligi meselesi haline getirmistir.

Bu calismada **"Hangi demografik ve davranissal faktorler oyun bagimlilik riskini en guclu bicimde yordamaktadir?"** sorusu merkeze alinmis; riskli bireylerin veri temelli olarak siniflandirilmasi hedeflenmistir.


## 1.4 Arastirma Sorulari

Bu calisma kapsaminda asagidaki arastirma sorulari belirlenmistir:

**Soru 1:** Gunluk oyun oynama suresi ile bireylerin uyku kalitesi arasinda anlamli bir iliski var midir?  
Oyun suresi arttikca uyku bozukluklarinin artip artmadigini anlamak, psikolojik etkinin temel gostergelerinden birini ortaya koyacaktir.

**Soru 2:** Oynanan oyun turu bireylerin sosyal izolasyon duzeyini etkiler mi?  
Farkli oyun turlerinin sosyal hayat uzerindeki etkisini belirlemek, hangi oyun kategorilerinin daha fazla risk tasidigi hakkinda fikir verecektir.

**Soru 3:** Ruh hali ve yoksunluk belirtileri gibi psikolojik gostergeler oyun bagimlilik riski ile iliski icinde midir?  
Bu degiskenler modele dahil edilmemis olsa da psikolojik etkinin gorsel analiz ile incelenmesi bagimlilik riskinin anlasilmasina katki sunacaktir.

**Soru 4:** Makine ogrenmesi algoritmalari kullanilarak bireylerin oyun bagimlilik riski basarili bicimde tahmin edilebilir mi?  
KNN, Random Forest ve Lojistik Regresyon algoritmalarinin karsilastirmali degerlendirilmesi en uygun modelin belirlenmesine olanak taniyacaktir.

# 3. VERI ON ISLEME (DATA PREPROCESSING)


## 3.1 Eksik Veri Temizleme


```python
df.isnull().sum()
```

```python
df["grades_gpa"] = df["grades_gpa"].fillna(df["grades_gpa"].mean())
df["work_productivity_score"] = df["work_productivity_score"].fillna(df["work_productivity_score"].mean())
```

Eksik veriler `isnull().sum()` ile dogrulandiktan sonra islenmistir. `grades_gpa` ve `work_productivity_score` degiskenlerindeki eksik degerler, ilgili sutunlarin **aritmetik ortalamasiyla** doldurulmustur. Ortalama atama yontemi; veri kaybini onlerken genel dagilimi koruma acisindan yaygin ve kabul gormis bir yaklasimdir. Islem sonrasinda veri setinde eksik gozlem kalmamistir.


## 3.2 Aykirik Deger Analizi (Outlier Detection)


```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=90)
plt.show()
```

```python
# IQR yöntemi ile aykırı değerleri baskılama
# Hedef değişken clipping'e dahil edilmemeli
clip_cols = [c for c in numerical_cols if c != "gaming_addiction_risk_level"]

Q1 = df[clip_cols].quantile(0.25)
Q3 = df[clip_cols].quantile(0.75)
IQR = Q3 - Q1

for col in clip_cols:
    alt_sinir = Q1[col] - 1.5 * IQR[col]
    ust_sinir = Q3[col] + 1.5 * IQR[col]
    df[col] = df[col].clip(alt_sinir, ust_sinir)
```

Sayisal degiskenlerdeki aykirik degerler iki asamada ele alinmistir. Ilk asamada **boxplot** grafikleriyle uc degerler gorsellestirilmistir. Ikinci asamada **IQR (Interquartile Range)** yontemi kullanilmis; her degisken icin alt ve ust sinirlar hesaplanmistir:

> Alt sinir = Q1 - 1.5 x IQR  
> Ust sinir = Q3 + 1.5 x IQR

Bu sinirlarin disinda kalan degerler veri setinden tamamen cikarilmak yerine **clipping** (baskilama) yontemiyle sinir degerlere cekilmistir. Boylece veri kaybi onlenmis, model performansinin daha kararli olmasi saglanmistir. Hedef degisken `gaming_addiction_risk_level` bu islemin disinda tutulmustur.

## 3.5 Veri Donusturme (Encoding)


**Kategorik Degiskenlerin Sayisallastirilmasi**


```python
from sklearn.preprocessing import LabelEncoder

# Encode oncesi hedef degisken etiketlerini sakla
y_raw = df["gaming_addiction_risk_level"].copy()

# Hedef değişkeni (target) kategorik özellikler listesinden çıkarıyoruz!
categorical_cols = [c for c in df.select_dtypes(include=['object']).columns if c != "gaming_addiction_risk_level"]

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

Makine ogrenmesi algoritmalari kategorik verileri dogrudan isleyemedigi icin, `object` tipindeki tum degiskenler **Label Encoding** yontemiyle sayisal degerlere donusturulmustur. Her sutun icin bagimsiz bir `LabelEncoder` nesnesi olusturulmus; bu sayede sutunlar arasinda encoding bilgisinin karismasinin onune gecilmis ve ileride gerektiginde tersine donusum yapilabilmesi saglanmistir.

**Not:** Nominal degiskenler icin One-Hot Encoding daha uygun bir alternatif olmakla birlikte, bu calismada model performansi ve boyut yonetimi acisindan Label Encoding tercih edilmistir.

## 4.6 Istatistiksel Karsilastirma - McNemar Testi


Modeller arasindaki performans farklarinin istatistiksel anlamliligsini test etmek amaciyla **McNemar testi** uygulanmistir. Bu test, ayni test kumesi uzerinde calisan iki siniflandirma modelinin hata matrislerini karsilastiran, bagimli orneklemler icin tasarlanmis bir yontemdir.

**Neden McNemar?** Modeller ayni veri seti uzerinde egitilip test edildigi icin gozlemler bagimsiz degildir; bu nedenle bagimli ornek testi olan McNemar tercih edilmistir. Bagimsiz gruplar icin kullanilan ki-kare testi burada uygun olmayacaktir.

**Hipotezler:**
- H0: Iki modelin hata oranlari arasinda istatistiksel olarak anlamli bir fark yoktur.
- H1: Iki modelin hata oranlari arasinda istatistiksel olarak anlamli bir fark vardir.
- Anlamlilik duzeyi: a = 0.05


### 4.6.1 Random Forest - Lojistik Regresyon Karsilastirmasi


```python
from statsmodels.stats.contingency_tables import mcnemar
import numpy as np

tb = np.zeros((2, 2))

for i in range(len(y_test)):
    rf_correct  = (y_pred_rf[i]  == y_test.iloc[i])
    log_correct = (y_pred_log[i] == y_test.iloc[i])

    if rf_correct and not log_correct:
        tb[0][1] += 1
    elif not rf_correct and log_correct:
        tb[1][0] += 1

result = mcnemar(tb)
print("p-değeri:", result.pvalue)
```

McNemar testi sonucunda **p = 0.337** bulunmustur. p > 0.05 oldugu icin H0 reddedilemez. Random Forest ile Lojistik Regresyon arasinda istatistiksel olarak **anlamli bir fark yoktur.**


### 4.6.2 KNN - Lojistik Regresyon Karsilastirmasi


```python
tb = np.zeros((2,2))

for i in range(len(y_test)):
    knn_correct = (y_pred_knn[i] == y_test.iloc[i])
    log_correct = (y_pred_log[i] == y_test.iloc[i])

    if knn_correct and not log_correct:
        tb[0][1] += 1
    elif not knn_correct and log_correct:
        tb[1][0] += 1

result = mcnemar(tb)
print("p-değeri:", result.pvalue)
```

McNemar testi sonucunda **p = 0.0004** bulunmustur. p < 0.05 oldugu icin H0 reddedilir. KNN ile Lojistik Regresyon arasinda istatistiksel olarak **anlamli bir fark vardir.** Lojistik Regresyon belirgin sekilde daha basarilidir.


### 4.6.3 KNN - Random Forest Karsilastirmasi


```python
tb = np.zeros((2,2))

for i in range(len(y_test)):
    knn_correct = (y_pred_knn[i] == y_test.iloc[i])
    rf_correct = (y_pred_rf[i] == y_test.iloc[i])

    if knn_correct and not rf_correct:
        tb[0][1] += 1
    elif not knn_correct and rf_correct:
        tb[1][0] += 1

result = mcnemar(tb)
print("p-değeri:", result.pvalue)
```

McNemar testi sonucunda **p = 0.005** bulunmustur. p < 0.05 oldugu icin H0 reddedilir. KNN ile Random Forest arasinda istatistiksel olarak **anlamli bir fark vardir.** Random Forest belirgin sekilde daha basarilidir.


**McNemar Testi Genel Sonucu**

|---|---|---|

RF ve Lojistik Regresyon birbirine istatistiksel olarak esit performans sergilemistir. KNN ise her iki modelden de istatistiksel olarak anlamli sekilde daha dusuk performans gostermistir.

## 4.10 Istatistiksel Karsilastirma - Wilcoxon Signed-Rank Testi


Modeller arasindaki performans farkinin istatistiksel olarak anlamli olup olmadigini belirlemek icin **Wilcoxon isaretli-siralama testi** uygulanmistir.

McNemar testi tek bir test seti uzerinde iki modeli karsilastirirken, Wilcoxon testi 30 farkli veri bolunmesinden elde edilen dogruluk dizilerini karsilastirmaktadir. Modeller ayni veri uzerinde tekrarli olarak egitildiginden olcumler bagimlidir; bu nedenle bagimli orneklemler icin uygun olan Wilcoxon testi tercih edilmistir.

**Hipotezler:**
- H0: Iki modelin dogruluk dagilimi arasinda istatistiksel olarak anlamli bir fark yoktur.
- H1: Iki modelin dogruluk dagilimi arasinda istatistiksel olarak anlamli bir fark vardir.
- Anlamlilik duzeyi: a = 0.05


```python
from scipy.stats import wilcoxon

stat, p = wilcoxon(dogruluk_rf, dogruluk_log)
print('RF vs Lojistik Reg. - Istatistik=%.3f, p=%.4f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Ayni dagilimdan ( H0 red edilmesi icin bir sebep bulunamadi)')
else:
    print('Farkli dagilimdan ( H0 red edildi)')
```

```python
stat, p = wilcoxon(dogruluk_knn, dogruluk_log)
print('KNN vs Lojistik Reg. - Istatistik=%.3f, p=%.6f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Ayni dagilimdan ( H0 red edilmesi icin bir sebep bulunamadi)')
else:
    print('Farkli dagilimdan ( H0 red edildi)')
```

```python
stat, p = wilcoxon(dogruluk_knn, dogruluk_rf)
print('KNN vs Random Forest - Istatistik=%.3f, p=%.6f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Ayni dagilimdan ( H0 red edilmesi icin bir sebep bulunamadi)')
else:
    print('Farkli dagilimdan ( H0 red edildi)')
```

Wilcoxon testi sonuclarina gore:

- **RF vs Lojistik Regresyon:** p = 0.380 > 0.05 → H0 reddedilemedi. Iki model arasinda istatistiksel olarak **anlamli bir fark yoktur.** Her iki model pratik acisindan esit gucte kabul edilebilir.
- **KNN vs Lojistik Regresyon:** p = 0.000002 < 0.05 → H0 reddedildi. KNN ile Lojistik Regresyon arasinda istatistiksel olarak **anlamli bir fark vardir.** Lojistik Regresyon KNN'e gore belirgin sekilde ustundur.
- **KNN vs Random Forest:** p = 0.000002 < 0.05 → H0 reddedildi. KNN ile Random Forest arasinda istatistiksel olarak **anlamli bir fark vardir.** Random Forest KNN'e gore belirgin sekilde daha basarilidir.

Genel degerlendirme: **Random Forest ve Lojistik Regresyon istatistiksel olarak birbirine esit**, ancak her ikisi de **KNN'den istatistiksel olarak anlamli sekilde ustundur.**
